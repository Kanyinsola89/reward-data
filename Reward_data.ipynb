```python
data = pd.read_excel('Cleaned_Mentorship_data2.xlsx')
```

```python
data.sort_values(['Mentor_Name', 'Session_Date'], inplace= True)
```

```python
#change the data type for the date column 
data['Session_Date'] = data['Session_Date'].dt.date
```

```python
data
```

```python
data['Points_Awarded'] = 0  # Initialize to 0

# Allocate initial points only once per mentor
def allocate_initial_points(row):
    if row['Mentor_Name'] not in allocated_points:
        allocated_points.add(row['Mentor_Name'])
        return 250
    return 0

allocated_points = set()
data['Points_Awarded'] = data.apply(allocate_initial_points, axis=1)

data.sort_values(['Mentor_Name', 'Session_Date'], inplace= True)

# Calculate points based on session duration and number
def calculate_session_points(row):
    if row['Session_Duration_Min'] >= 30:
        if row['Session_Number'] == 1:
            return row['Points_Awarded'] + 250
        elif row['Session_Number'] == 2:
            return row['Points_Awarded'] + 500
    return row['Points_Awarded']

data['Points_Awarded'] = data.apply(calculate_session_points, axis=1)

# Group by mentor and sum up points
data['Total_Points'] = data.groupby('Mentor_Name')['Points_Awarded'].transform('sum')

# Print the final DataFrame
print(data)
```
